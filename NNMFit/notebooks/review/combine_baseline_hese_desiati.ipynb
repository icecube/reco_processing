Lets make the plot paolo requested

```python
def plot_histogram_double(hist_graph_hdl, det_config, input_variables, title=None, savepath=None,
                          hist_graph_hdl2=None, hdl1_label="hdl1", hdl2_label="hdl2"):

    print(det_config)
    binnings = hist_graph_hdl.get_binning(det_config=det_config)

    fig, axes = plt.subplots(3, 2, figsize=(10, 12), gridspec_kw={"height_ratios": [3, 3, 1]})
    axes_linear = axes[0]
    axes_log    = axes[1]
    axes_ratio  = axes[2]

    handlers = [(hist_graph_hdl, "solid"), (hist_graph_hdl2, "dashed")] if hist_graph_hdl2 is not None else [(hist_graph_hdl, "solid")]
    colors = [f"C{i}" for i in range(len(input_variables))]

    stored = {j: {} for j in range(len(input_variables))}

    for hdl_idx, (hdl, linestyle) in enumerate(handlers):
        for j, ((input_name, input_variable), color) in enumerate(zip(input_variables.items(), colors)):

            res = hdl.get_evaled_histogram(input_variables=input_variable, det_config=det_config, reshape=True)

            for i, (variable_name, binning) in enumerate(binnings.items()):

                hist = res["mu"].sum(axis=1 - i)
                yerror = np.sqrt(res["ssq"].sum(axis=1 - i))

                if i not in stored[j]:
                    stored[j][i] = {}
                stored[j][i][hdl_idx] = (hist, yerror, binning)

                label = f"{input_name}: {sum(hist):.2f}+-{sum(yerror):.2f}" if i == hdl_idx else None

                for ax_row in [axes_linear, axes_log]:
                    ax_row[i].stairs(hist, binning, label=label, linestyle=linestyle, color=color)
                    ax_row[i].fill_between(
                        binning,
                        np.r_[hist - yerror, (hist - yerror)[-1]],
                        np.r_[hist + yerror, (hist + yerror)[-1]],
                        step="post", alpha=0.2, color=color
                    )
                    ax_row[i].set_xlim(min(binning), max(binning))
                    ax_row[i].set_ylabel(f"Rate / {hdl.get_livetime(det_config) / (3600 * 24 * 365.25):.2f} yr")
                    if "energy" in variable_name or "length" in variable_name:
                        ax_row[i].set_xscale("log")
                        axes_ratio[i].set_xscale("log")

                axes_log[i].set_yscale("log")

    # Legends
    legend_titles = [hdl1_label, hdl2_label] if hist_graph_hdl2 is not None else [hdl1_label]
    for i, legend_title in enumerate(legend_titles):
        axes_linear[i].legend(title=legend_title, fontsize=8, title_fontsize=9)
        axes_log[i].legend(title=legend_title, fontsize=8, title_fontsize=9)

    # Ratio plots
    for j, ((input_name, input_variable), color) in enumerate(zip(input_variables.items(), colors)):
        for i, (variable_name, binning) in enumerate(binnings.items()):
            if hist_graph_hdl2 is not None and 0 in stored[j][i] and 1 in stored[j][i]:
                hist1, yerr1, binning = stored[j][i][0]
                hist2, yerr2, _ = stored[j][i][1]

                with np.errstate(divide="ignore", invalid="ignore"):
                    ratio = np.where(hist2 != 0, hist1 / hist2, np.nan)
                    ratio_err = np.where(
                        hist2 != 0,
                        ratio * np.sqrt((yerr1 / np.where(hist1 != 0, hist1, np.nan)) ** 2 +
                                        (yerr2 / np.where(hist2 != 0, hist2, np.nan)) ** 2),
                        np.nan
                    )

                axes_ratio[i].stairs(ratio, binning, color=color, linestyle="solid")
                axes_ratio[i].fill_between(
                    binning,
                    np.r_[ratio - ratio_err, (ratio - ratio_err)[-1]],
                    np.r_[ratio + ratio_err, (ratio + ratio_err)[-1]],
                    step="post", alpha=0.2, color=color
                )

    for i, (variable_name, binning) in enumerate(binnings.items()):
        axes_ratio[i].axhline(1.0, color="gray", linestyle="--", linewidth=0.8, zorder=0)
        axes_ratio[i].set_xlabel(x_labels[variable_name])
        axes_ratio[i].set_ylabel(f"{hdl1_label} / {hdl2_label}")
        axes_ratio[i].set_xlim(min(binning), max(binning))

    # Remove x tick labels from linear and log rows
    for ax in [*axes_linear, *axes_log]:
        ax.set_xticklabels([])

    plt.suptitle(det_config if not title else title)
    plt.tight_layout()
    if savepath:
        plt.savefig(savepath)
    plt.show()
```
