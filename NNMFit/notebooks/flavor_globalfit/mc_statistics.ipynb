```python
configs_dir = "/data/user/tvaneede/GlobalFit/reco_processing/NNMFit/configs/flavor_globalfit/"
config_hdl = AnalysisConfig.from_configs(
        main_config_file=f"{configs_dir}/main.cfg",
        analysis_config_file = f"{configs_dir}/analysis_configs/asimov/SAY/asimov_SAY_HESEBestfit.yaml",
        config_dir=configs_dir,
        override_dict=None,
        override_config_files=["override/systematics/NoSystematics.cfg", "override/datasets_hese/mcd-simpletopology_flux-hese_feat-11features_plus_rloglmilli_econf_evtgen/bdt1_0.3_bdt2_0.4_length_10.cfg", "override/binning/hese/10bdtprod_threshold_0.12.cfg"],
        override_components_files=None,
        override_parameters_files=None)
# build a histogram
hist_graph_hdl = HistogramGraph(config_hdl)
detector_configs = config_hdl.get_det_configs()
```

```python
plotting_path_hist = f"{plotting_path}/11features_plus_rloglmilli_econf_evtgen_bdtprod"
os.system(f"mkdir -p {plotting_path_hist}")
from plot_utils import *
for det_config in detector_configs: 
    plot_histogram_2D(hist_graph_hdl=hist_graph_hdl, det_config=det_config, input_variable = input_variables["All"], zlog=False, 
                      savepath=f"{plotting_path_hist}/hist_2D_{det_config}.png")
    plot_histogram_flavor_2D(hist_graph_hdl=hist_graph_hdl, det_config=det_config, gamma_astro=2.87, astro_norm=2.1233, zlog=None, 
                        savepath=f"{plotting_path_hist}/hist_2D_{det_config}_flavor.png")
    plot_2d_ssq_and_relerr(hist_graph_hdl=hist_graph_hdl, det_config=det_config, input_variable = input_variables["All"], zlog=False, 
                      savepath=f"{plotting_path_hist}/ssq_relerr_{det_config}.png")
    plot_2d_relerr(hist_graph_hdl=hist_graph_hdl, det_config=det_config, input_variable = input_variables["All"], zlog=False, 
                      savepath=f"{plotting_path_hist}/relerr_{det_config}.png")
```
