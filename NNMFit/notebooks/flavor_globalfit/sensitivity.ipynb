```python
plotting_path = "/data/user/tvaneede/GlobalFit/reco_processing/NNMFit/notebooks/flavor_globalfit/sensitivity"
os.system(f"mkdir -p {plotting_path}")
```

```python
# load the data
drop_unsuccessful = False
scan_dict = {}
data = {}
for name, scan_dir in scan_dir_dict.items():
    print(name)
    scan_dict[name] = ScanHandler(scan_dir,dump=False,drop_unsuccessful=drop_unsuccessful)
    data[name] = {}
    data[name]["ftau_asimov_poisson"],data[name]["fe_asimov_poisson"],data[name]["ft_grid_asimov_poisson"],data[name]["fe_grid_asimov_poisson"],data[name]["ts_grid_asimov_poisson"] = get_contour_points(scan_dict,name)
```

```python
# spice_ftp_oldpid_bdtprod
names = [
"spice_nosyst",
"hese_oldpid_nosyst",
"hese_allfeat_bdtprod_combinedBase",
]
labels = [
    "Spice, old reco",
    "FTP, new reco",
    "FTP, BDT",
]
C = compare_contours( data = data,
                  names = names,
                  labels = labels,
                  levels = ["68%"],
                  title = r"$\phi_0 = 2.12,\gamma=2.87$",
                  savepath = f"{plotting_path}/spice_ftp_oldpid_bdtprod.pdf")

for name in names: compare_area( C, alt_key=name, base_key="spice_nosyst", levels=["68%"] )
```
