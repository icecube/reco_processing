```python
for cfg in CONFIGS:
    name = cfg["name"]
    outpath = os.path.join(output_dir, name)
    outfile = os.path.join(outpath, "Precalculated_Graph.pickle")
    os.makedirs(outpath, exist_ok=True)

    NNMFit_config_options = {
        "init_method": "from_configs",
        "main_config": os.path.join(configs_dir, "main.cfg"),
        "analysis_config": os.path.join(configs_dir, cfg["analysis_config"]),
        "config_dir": configs_dir,
        "override_configs": cfg["override_configs"],
        "override_components": cfg["override_components"],
        "override_parameters": cfg["override_parameters"],
    }

    cmd = [
        python, script,
        "--main_config", os.path.join(configs_dir, "main.cfg"),
        "--analysis_config", os.path.join(configs_dir, cfg["analysis_config"]),
        "--config_dir", configs_dir,
        "--override_configs", *cfg["override_configs"],
        "--override_components", *cfg["override_components"],
        "-o", outfile,
    ]
    if cfg["override_parameters"]:
        cmd += ["--override_parameters", *cfg["override_parameters"]]

    result = subprocess.run(cmd, capture_output=True, text=True)
    print(result.stdout)
    if result.stderr:
        print(result.stderr)

    make_histogram_from_variables(NNMFit_config_options=NNMFit_config_options, out_file=f"{outpath}/MC_Histogram.pickle")
    generate_data_hist(scan_path=outpath)
```
